```python
f = "csv/all_qualifications.csv"
df = pd.read_csv(f, dtype=str)
print("...rows:", df.shape[0])
df
```

```python
# Remove prefix from some files.

import glob
import os, re

for p in ["pdf", "png"]:

    for old in glob.glob(os.path.join(p, f"*.{p}")):
    
        new = list(os.path.split(old))
        new[-1] = new[-1].removeprefix("CertificateOfCompletion_").lower()
        new[-1] = re.sub("\\s+", "_", new[-1])
        new = os.path.join(*new)
        if old != new:
            print(old, " --> ", new)
            os.rename(old, new)
```

```python
# remove prefix dir in dataframe

df["pdf"] = df["pdf"].apply(lambda x: x if type(x)!=str else x.removeprefix("pdf/"))
df["png"] = df["png"].apply(lambda x: x if type(x)!=str else x.removeprefix("png/"))
df.to_csv(f, index=False)
```

```python
# Check for missing files (in csv but not in file system).

for index, row in df.iterrows():
    for file_type in ["pdf", "png"]:
        fn = str(row[file_type])
        if fn != "nan":
            fp = os.path.join(file_type, fn)
            if not os.path.exists(fp):
                print(f"...{file_type} file not found:", fp)
```

```python
#  Check for extra files (in file system but not in csv).

root_dir = "."
for dirpath, dirnames, filenames in os.walk(root_dir):

    for file_type in ["pdf", "png"]:
        if dirpath  == os.path.join(".", file_type):
            for f in filenames:
                if not f in df[file_type].values:
                    ans = input(f"Delete file {f}? [Y/N] ")
                    if ans == "Y":
                        fp = os.path.join(".", file_type, f)
                        os.remove(fp)
```

```python
import string, re

def make_id(q):
    a = q.lower()
    a = re.sub('\\s+', '-', a)
    a = re.sub('[^0-9a-zA-Z-]+', '', a)
    return a

def make_link(q):
    return f"[{q}](#{make_id(q)})"

def make_bold(q):
    str_ = q
    if "associate" in q.lower():
        return f"**{q}**"
    if "specialization" in q.lower():
        return f"**{q}**"    
    return str_
    
df = df.loc[:, ["qualification", "institution", "date"]]
df["qualification"] = df["qualification"].apply(make_link)
df["qualification"] = df["qualification"].apply(make_bold)

df.columns = [string.capwords(c) for c in df.columns]
str_ = df.to_markdown(index=False)
print(str_)
```
