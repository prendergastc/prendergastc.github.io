```python
df = pd.read_csv(f)
df
```

```python
df.columns = [c.lower() for c in df.columns]

for index, row in df.iterrows():

    d = str(row["date"])
    png = str(row["png"])
    pdf = str(row["pdf"])
    t = str(row["title"])

    stem = "images/esri_certificates/"
    
    image = (
        '[!'
            + '[' + t + ']'
            + '(' + stem + png + ')'
            + '{: .image-process-large-photo}'
        + ']'
        + '(' + stem + png + ')'
        + '{class="mybox"}'
    )


    lst = []

    print("###", t)
    print()
    print(image)
    print()

    if pdf.strip() != "nan":
        p = '[PDF file]({static}' + stem + pdf + ')'
        lst.append(p)
    
    lst.append("[Return to overview](#overview)")
    print(" | ".join(lst))
    print()
```

```python
import string, re

def make_id(q):
    a = q.lower()
    a = re.sub('\\s+', '-', a)
    a = re.sub('[^0-9a-zA-Z-]+', '', a)
    return a

def make_link(q):
    return f"[{q}](#{make_id(q)})"
    
df = df.loc[:, ["title", "date"]]
df["title"] = df["title"].apply(make_link)

df.columns = [string.capwords(c) for c in df.columns]
str_ = df.to_markdown(index=False)
print(str_)
```
